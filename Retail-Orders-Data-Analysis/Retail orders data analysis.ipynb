## Read csv file and handle null values

```python
import pandas as pd
import numpy as np
```

```python
# read data from csv file
df = pd.read_csv('orders.csv')
df.head(20)
```

```python
df.isnull().sum()
```

```python
df['Ship Mode'].unique()
```

```python
df['Ship Mode'] = df['Ship Mode'].replace({'Not Available':np.nan,'unknown':np.nan})
```

```python
df['Ship Mode'].unique()
```

```python
df.isnull().sum()
```

```python
# another method to read file and handle null values
df = pd.read_csv('orders.csv', na_values=['Not Available','unknown'])
df['Ship Mode'].unique()
```

```python
df.isnull().sum()
```

## Rename column names

```python
# rename column names to lower case and space with underscore
df.columns
```

```python
# 1st method using for loop and list methods
column_list = []
for col in df.columns:
    column_list.append(col.replace(' ','_').lower())

df.columns = column_list
df.columns
```

```python
# 2nd method to rename columns
df.columns = df.columns.str.lower().str.replace(' ','_')
df.columns
```

```python
# 3rd method using list comprehensions
df.columns = [col.replace(' ','_').lower() for col in df.columns]
df.columns
```

## Derive new columns discount, sale price and pofit

```python
df.head()
```

```python
df['discount'] = df['list_price'] * df['discount_percent'] / 100
df['sale_price'] = df['list_price'] - df['discount']
df['profit'] = df['sale_price'] - df['cost_price']
df.head()
```

## Drop unnecessary columns - cost_price, list_price, discount_percent

```python
df.drop(columns=['cost_price', 'list_price', 'discount_percent'], inplace=True)
df.head()
```
